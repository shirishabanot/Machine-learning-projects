```python
import os

def read_text_files(output_dir_path):
    files_data = []
    for root, dirs, files in os.walk(output_dir_path):
        print(f"Reading Text Files from {root}... Detected {len(files)} files in {root} directory.")
        for file_name in files:
            if file_name.endswith(".txt"):
                file_path = os.path.join(root, file_name)
                with open(file_path, "r", encoding="utf-8") as f:  # Specify encoding="utf-8"
                    file_content = f.read()
                    lines = file_content.splitlines()
                    if len(lines) >= 9:
                        file_data = {
                            "ProductId": lines[0],
                            "UserId": lines[1],
                            "ProfileName": lines[2],
                            "HelpfulnessNumerator": lines[3],
                            "HelpfulnessDenominator": lines[4], 
                            "Score": lines[5], 
                            "Time": lines[6],
                            "ReviewSummary": lines[7], 
                            "ReviewText": lines[8] 
                        }
                        files_data.append(file_data)
                    else:
                        print(f"File {file_name} does not contain enough lines of data.")
    return files_data

output_dir_path = r"C:\Users\hp\Desktop\Text_Project"
files_data = read_text_files(output_dir_path)

```

```python
df = pd.DataFrame(files_data)
print(df)
```

```python
for column in ['ProductId', 'UserId','ProfileName',
       'HelpfulnessNumerator', 'HelpfulnessDenominator', 'Score', 'Time',
       'ReviewSummary', 'ReviewText']:
    df[column] = df[column].apply(lambda x: x.split(': ')[1])

# Display the cleaned DataFrame
df
```

```python
df['HelpfulnessNumerator'] = pd.to_numeric(df['HelpfulnessNumerator'])
df['HelpfulnessDenominator'] = pd.to_numeric(df['HelpfulnessDenominator'])

# Calculate Helpfulness
df['Helpfulness'] = df.apply(
    lambda row: row['HelpfulnessNumerator'] / row['HelpfulnessDenominator'] if row['HelpfulnessDenominator'] != 0 else 0, 
    axis=1
)

# Display the DataFrame with the new 'Helpfulness' column
df
```

```python
def score_review(score):
    if score==1 or score==2:
        return 'Negative'
    elif score==3:
        return "Neutral"
    else:
        return "Positive"
```

```python
df['Sentiment']=df['Score'].apply(score_review)
```
