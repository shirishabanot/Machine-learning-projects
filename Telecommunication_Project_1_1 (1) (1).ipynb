## Difference in churn between senior citizens and non-senior citizens

```python
churn_senior = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

colors = ['#beb9db', '#fdcce5']  

plt.figure(figsize=(8, 6))
plt.pie(churn_senior.sum(axis=1), labels=['Non-Senior Citizen', 'Senior Citizen'], autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Churn Rate Variation Between Senior and Non-Senior Citizens')
plt.show()
```

## Checking whether having a partner or dependents impact churn

```python
colors = ['#beb9db', '#fdcce5']

# Calculate churn counts based on Partner and Dependents
churn_partner_dependents = df.groupby(['Partner', 'Dependents', 'Churn']).size().unstack()

fig, ax = plt.subplots(figsize=(10, 6))
churn_partner_dependents.plot(kind='bar', ax=ax, color=colors)

plt.title('Churn Rate Based on Partner and Dependents')
plt.xlabel('(Partner, Dependents)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Churn', labels=['No', 'Yes'])
plt.show()
```

### Partner status has a small impact on churn rate with dependents: 
* The churn rate is slightly higher for customers with partners than for those without partners, regardless of whether they have dependents or not.  However, the difference in churn rate between partners and non-partners seems relatively small for both groups with and without dependents.

### Dependents have a more significant impact on churn rate: 
* The churn rate is significantly lower for customers with dependents compared to those without dependents, regardless of their partner status. This suggests that having dependents makes customers less likely to churn.

### The impact of dependents is independent of partner status: 
* The gap in churn rate between customers with dependents and those without dependents is similar for both partner groups. This suggests that having dependents is a stronger indicator of lower churn than whether a customer has a partner.

## Does the presence of online security, device protection, or tech support influence churn

```python
df2 = df.copy()

# Convert categorical columns to numeric
df2['StreamingTV'] = df2['StreamingTV'].map({'Yes': 1, 'No': 0})
df2['StreamingMovies'] = df2['StreamingMovies'].map({'Yes': 1, 'No': 0})
df2['Churn'] = df2['Churn'].map({'Yes': 1, 'No': 0})

# Group by StreamingTV and StreamingMovies, then sum the Churn column
churn_streaming = df2.groupby(['StreamingTV', 'StreamingMovies'])['Churn'].sum()

# Plotting the churn rate based on StreamingTV and StreamingMovies with a pie chart
labels = ['No TV, No Movies', 'TV, No Movies', 'No TV, Movies', 'TV, Movies']
colors = ['#beb9db', '#fdcce5', '#2C4E80', '#E1AFD1']  # Custom colors for the pie chart

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(churn_streaming, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Churn Rate Variation Based on Streaming Services')
plt.show()
```

### Customers with multiple streaming service subscriptions churn less:
* The pie chart shows the smallest slice is for customers who subscribe to both TV and Movie services, suggesting this group has the lowest churn rate.

### Customers with no streaming services churn the most:
* The largest slice of the pie chart is for customers who subscribe to no streaming services (No TV, No Movies). This indicates that this group has the highest churn rate.

## Does the payment method affect churn behavior

```python
churn_payment = df.groupby(['PaymentMethod', 'Churn']).size().unstack()
churn_payment['Churn Rate'] = churn_payment['Yes'] / (churn_payment['Yes'] + churn_payment['No']) * 100

churn_payment['Churn Rate'].plot(kind='bar', figsize=(10, 6), color='#2C4E80')
plt.title('Churn Rate by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Automatic bank transfer has the lowest churn rate: 
* The bar graph shows the churn rate for bank transfer (automatic) is the lowest compared to all other payment methods. This suggests that customers who use automatic bank transfer are less likely to churn than those using other payment methods.

### Credit card with automatic payment has a lower churn rate than credit card with manual payment:
* The churn rate for credit card (automatic) is lower than the churn rate for credit card (manual). This suggests that setting up automatic payments can reduce churn among credit card users.

### Electronic check and mailed check have similar churn rates, and the highest overall: 
* The churn rates for electronic check and mailed check are the highest among all the payment methods and are relatively close in value. This suggests that both electronic check and mailed check users have a higher likelihood of churning compared to other payment methods.

## By taking into consideration above factors data shows a clear benefit to churn reduction for automatic payment methods

## Correlation between monthly charges, total charges, and churn

```python
data = df[['MonthlyCharges', 'TotalCharges', 'Churn']]

# Convert Churn to numeric (1 for Yes, 0 for No) using .loc
data.loc[:, 'Churn'] = data['Churn'].map({'Yes': 1, 'No': 0})

correlation_matrix = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix: Monthly Charges, Total Charges, and Churn')
plt.show()
```

### Strong positive correlation between monthly charges and churn:
* The heat map shows a dark red square between "Monthly Charges" and "Churn," indicating a strong positive correlation. This means that as monthly charges increase, churn also tends to increase. Customers who pay more per month are more likely to churn.

### Weaker positive correlation between total charges and churn: 
* The heat map shows a light orange square between "Total Charges" and "Churn." This indicates a weaker positive correlation compared to monthly charges. While there is a positive relationship, the increase in churn rate is less pronounced as total charges rise compared to monthly charges.

### Monthly charges stronger indicator for churn than total charges:  
* By comparing the color intensity of the two squares, we can see that the correlation between monthly charges and churn is stronger than the correlation between total charges and churn. This suggests that customers are more sensitive to recent (monthly) charges than the total amount they have paid overall.

## Coclusion

* Overall, factors such as having dependents, longer tenure, phone service, multiple streaming service subscriptions, and using automatic payment methods contribute to lower churn rates.
* On the other hand, factors like higher monthly charges, using electronic or mailed checks for payment, and being a non-senior citizen are associated with higher churn rates.
* Customer satisfaction and loyalty programs targeting specific demographics (like senior citizens, customers with dependents, etc.) can be effective strategies to reduce churn and improve overall customer retention.

```python
X = df.drop(columns=["Churn","customerID"])
```

```python
y = df["Churn"]
```

## Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 0)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
X_train.head()
```

```python
X_train_cat = X_train.drop(columns = ['tenure','MonthlyCharges','TotalCharges'])
X_train_num = X_train.select_dtypes(include = ["int",'float64'])
print(X_train_num.shape)
print(X_train_cat.shape)
```

```python
X_train_cat.columns
```

```python
X_train_cat.head()
```

```python
X_test_cat = X_test.drop(columns = ['tenure','MonthlyCharges','TotalCharges'])
X_test_num = X_test.select_dtypes(include = ['int64','float64'])
print(X_test_cat.shape)
print(X_test_num.shape)
```

## Applying Normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_num_transformed = pd.DataFrame(scaler.fit_transform(X_train_num),
                                       columns = scaler.get_feature_names_out(),
                                       index = X_train_num.index)
```

```python
X_test_num_transformed = pd.DataFrame(scaler.transform(X_test_num),
                                     columns = scaler.get_feature_names_out(),
                                     index = X_test_num.index)
```

## Applying One hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
```

```python
X_train_cat_transformed = pd.DataFrame(ohe.fit_transform(X_train_cat),
                                       columns = ohe.get_feature_names_out(),
                                       index = X_train_cat.index)
```

```python
X_test_cat_transformed = pd.DataFrame(ohe.transform(X_test_cat),
                                       columns = ohe.get_feature_names_out(),
                                       index = X_test_cat.index)
```

## Concate

```python
X_train_transformed = pd.concat([X_train_num_transformed, X_train_cat_transformed], axis = 1)
X_train_transformed.shape
```

```python
X_test_transformed = pd.concat([X_test_num_transformed, X_test_cat_transformed], axis = 1)
X_test_transformed.shape
```
